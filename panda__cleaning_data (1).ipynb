```python
import pandas as pd

name=pd.DataFrame({
    "customer_id":[1,2,3,4,5,6,7],
    "purchase":[500,200,100,123,233,345,334]
})

id_df=pd.DataFrame({
    "id_name":[1,2,3,4,5,6,7],
    "name":["Ali","rana","jone","mike","smith","aliya","moh"]

})

merge = pd.merge(name,id_df, left_on="customer_id",right_on="id_name",how="left")
```

```python
merge.isnull()
```

```python
Purchase =merge[merge["purchase"]>300]
```

```python
sorted=merge.sort_values(by="purchase",ascending=False)
sorted
```

```python
sum = merge["purchase"].sum()
print()
```

```python
total_purchase=merge["purchase"].sum()
print("Total Purchase:",total_purchase)
```

```python
avg_purchase = merge["purchase"].mean()
print("avg_purchase", avg_purchase)
```
